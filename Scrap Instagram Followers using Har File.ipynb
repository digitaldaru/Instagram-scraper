```python
import json
import pandas as pd

# Read json file with utf-8 encoding
with open("www.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Create empty list to store indexes with '"users\"' in value
indexes = []

# Loop through range of len(log>entries)
for i in range(len(data["log"]["entries"])):
    # Try to find key with '"users\"' in value
    try:
        if '"users\"' in data["log"]["entries"][i]["response"]["content"]["text"]:
            # Append index number to list
            indexes.append(i)
    except KeyError:
        pass

# Create empty lists to store user information
usernames = []
full_names = []
is_privates = []
is_verifieds = []
profile_pic_urls = []

# Loop through indexes with '"users\"' in value
for index in indexes:
    # Load json from log>entries>index number>response>content>text
    user_data = json.loads(data["log"]["entries"][index]["response"]["content"]["text"])
    # Loop through key named as users
    for user in user_data["users"]:
        # Append user information to lists
        usernames.append(user["username"])
        full_names.append(user["full_name"])
        is_privates.append(user["is_private"])
        is_verifieds.append(user["is_verified"])
        profile_pic_urls.append(user["profile_pic_url"])

# Convert lists to data frame
df = pd.DataFrame({"Username": usernames, "Full Name": full_names, "Is Private": is_privates, "Is Verified": is_verifieds, "Profile Pic URL": profile_pic_urls})

df.head()
```

# Instagram Followers Scraper

```python
import json
import pandas as pd

# Read the JSON file with UTF-8 encoding
with open("www.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Find all the keys inside log>entries>range value>response>content>text
entries = data["log"]["entries"]
index_list = []
for i in range(len(entries)):
    try:
        response_text = entries[i]["response"]["content"]["text"]
        if '"users\"' in response_text:
            index_list.append(i)
    except:
        continue

# Load the text as JSON and append values to lists
username_list = []
full_name_list = []
is_private_list = []
is_verified_list = []
profile_pic_url_list = []
for index in index_list:
    response_text = entries[index]["response"]["content"]["text"]
    response_json = json.loads(response_text)
    for user in response_json["users"]:
        username_list.append(user["username"])
        full_name_list.append(user["full_name"])
        is_private_list.append(user["is_private"])
        is_verified_list.append(user["is_verified"])
        profile_pic_url_list.append(user["profile_pic_url"])

# Convert lists to a data frame
df = pd.DataFrame({"Username": username_list,
                   "Full Name": full_name_list,
                   "Is Private": is_private_list,
                   "Is Verified": is_verified_list,
                   "Profile Pic URL": profile_pic_url_list})
```
